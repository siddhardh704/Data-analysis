```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel(r"C:\Users\siddh\OneDrive\Desktop\siddhardh\sales_dataset.xlsx")
```

```python
print(df.head())
print(df.info())
print(df.shape)
```

```python
df = df.dropna()
```

```python
total_sales=df["TotalSales"].sum()
print("total sales",total_sales)
```

```python
high_sales=df[df["TotalSales"]>100]
print("highest sales",high_sales)
```

```python
df.groupby('Date')['TotalSales'].sum().plot(marker='o')
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
category_sales = df.groupby('Category')['TotalSales'].sum()
category_sales.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```
